# CLUSTERING USING K MEANS

```python
X = dtset.iloc[:, [0, 1, 2, 3]].values
```

```python
print(X)
```

```python
from sklearn.cluster import KMeans
```

```python
distance = []

## this loop for determining the ideal elbow

for k in range(1,11):
    km = KMeans(n_clusters=k)
    km = km.fit(X)
    distance.append(km.inertia_)
```

```python
plt.plot(range(1,11),distance)
plt.title(' ELBOW METHOD')
plt.xlabel('Number of Clusters')
plt.ylabel('distance')
plt.show()
```

 so the optimum no of clusters in this case is 3

```python
# applying k means to the dtset
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
# visualizing the clusters between first two columns:
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], 
            s = 90, c = 'red', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], 
            s = 90, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1],
            s = 90, c = 'green', label = 'Iris-virginica')
plt.legend()
```

```python
# visualizing the clusters between first and third columns:
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 2], 
            s = 90, c = 'red', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 2], 
            s = 90, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 2],
            s = 90, c = 'green', label = 'Iris-virginica')

plt.legend()
```

```python
# visualizing the clusters between first and fourth column:
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 3], 
            s = 90, c = 'red', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 3], 
            s = 90, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 3],
            s = 90, c = 'green', label = 'Iris-virginica')


plt.legend()
```
